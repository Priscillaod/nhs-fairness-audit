```python
#03. Model Development
This notebook implements the model development phase described in Chapter 4.2  of the dissertation titled:
"Fairness and Explainability Audits in NHS Predictive Models using Synthetic Data."
```

```python
# Import libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.linear_model import LogisticRegression
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report
import matplotlib.pyplot as plt
import seaborn as sns

# Load both datasets generated in the previous step
df = pd.read_csv("synthetic_nhs_dataset.csv")
original_df = pd.read_csv("synthetic_nhs_original.csv")

# Display basic info
print("Dataset Loaded Successfully ✅")
print(df.shape)
df.head()
```

```python
#4.2.1 Data Preparation and Train–Test Split

Before model training, the dataset was preprocessed to ensure analytical consistency and to avoid data leakage.
```

```python
# Separate feature matrix (X) and target vector (y)
X = df[['age', 'gender', 'ethnicity', 'imd_quintile', 'comorbidity_score', 'discharge_destination']]
y = df['readmitted_30days']

# Define categorical and numerical columns
categorical_features = ['gender', 'ethnicity', 'discharge_destination']
numeric_features = ['age', 'imd_quintile', 'comorbidity_score']

# Preprocessing: OneHotEncoding + Scaling
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer

# One-hot encode categorical features and scale numeric ones
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numeric_features),
        ('cat', OneHotEncoder(drop='first', handle_unknown='ignore'), categorical_features)
    ])

# Apply transformation
X_prepared = preprocessor.fit_transform(X)

# Confirm shape and dimensionality
print("✅ Preprocessing complete")
print(f"Feature matrix shape: {X_prepared.shape}")

# Split into train and test sets
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_prepared, y, test_size=0.2, random_state=42, stratify=y)

print(f"Training set: {X_train.shape}, Test set: {X_test.shape}")
print(f"Positive class proportion (Train): {y_train.mean():.3f}")
print(f"Positive class proportion (Test): {y_test.mean():.3f}")
```

```python
#4.2.2: Logistic Regression (Baseline Model)
```

```python
# Step 1 — Train a Logistic Regression baseline model
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix, classification_report, RocCurveDisplay

# Initialise logistic regression with balanced class weights (to handle mild imbalance)
log_reg = LogisticRegression(max_iter=1000, solver='lbfgs', class_weight='balanced', random_state=42)

# Fit model
log_reg.fit(X_train, y_train)

# Predict on test set
y_pred = log_reg.predict(X_test)
y_pred_prob = log_reg.predict_proba(X_test)[:, 1]

# Evaluate baseline metrics
acc = accuracy_score(y_test, y_pred)
prec = precision_score(y_test, y_pred)
rec = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
roc_auc = roc_auc_score(y_test, y_pred_prob)

print("✅ Logistic Regression Baseline Results")
print(f"Accuracy:  {acc:.3f}")
print(f"Precision: {prec:.3f}")
print(f"Recall:    {rec:.3f}")
print(f"F1 Score:  {f1:.3f}")
print(f"ROC-AUC:   {roc_auc:.3f}")

# Plot ROC Curve
import matplotlib.pyplot as plt
RocCurveDisplay.from_estimator(log_reg, X_test, y_test)
plt.title("Figure 4.X – ROC Curve for Logistic Regression Baseline")
plt.show()
```

```python
#4.2.3: Advanced Model: XGBoost Classifier
```

```python
#Step 1: Import and Train the Model
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, RocCurveDisplay

# Initialise XGBoost model
xgb_model = XGBClassifier(
    n_estimators=300,
    learning_rate=0.05,
    max_depth=4,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42,
    scale_pos_weight=(1 / 0.31),  # balance positive class (~31%)
    use_label_encoder=False,
    eval_metric='logloss'
)

# Fit the model
xgb_model.fit(X_train, y_train)

# Predict
y_pred_xgb = xgb_model.predict(X_test)
y_pred_prob_xgb = xgb_model.predict_proba(X_test)[:, 1]

# Evaluate metrics
acc_xgb = accuracy_score(y_test, y_pred_xgb)
prec_xgb = precision_score(y_test, y_pred_xgb)
rec_xgb = recall_score(y_test, y_pred_xgb)
f1_xgb = f1_score(y_test, y_pred_xgb)
roc_auc_xgb = roc_auc_score(y_test, y_pred_prob_xgb)

print("✅ XGBoost Model Results")
print(f"Accuracy:  {acc_xgb:.3f}")
print(f"Precision: {prec_xgb:.3f}")
print(f"Recall:    {rec_xgb:.3f}")
print(f"F1 Score:  {f1_xgb:.3f}")
print(f"ROC-AUC:   {roc_auc_xgb:.3f}")

# Plot ROC curve
RocCurveDisplay.from_estimator(xgb_model, X_test, y_test)
plt.title("Figure 4.X – ROC Curve for XGBoost Model")
plt.show()
```

```python
# 4.3.2 FAIRNESS AUDIT USING FAIRLEARN

from fairlearn.metrics import MetricFrame, selection_rate, demographic_parity_difference, equalized_odds_difference
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

# Recreate matching test indices 
X_train_idx, X_test_idx = train_test_split(df.index, test_size=0.2, random_state=42, stratify=y)

# Match to the unencoded categorical version
X_test_df = original_df.iloc[X_test_idx]

# Define protected attributes
protected_features = {
    "Gender": X_test_df["gender"],
    "Ethnicity": X_test_df["ethnicity"],
    "IMD Quintile": X_test_df["imd_quintile"]
}
```

```python
# Quick sanity check
print("✅ Protected features defined successfully.")
print("X_test_df shape:", X_test_df.shape)
print("\nSample protected attributes:")
for k, v in protected_features.items():
    print(f"{k}: {v.unique()[:5]}")
```

```python
from fairlearn.metrics import (
    MetricFrame, 
    selection_rate, 
    demographic_parity_difference, 
    equalized_odds_difference
)
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import matplotlib.pyplot as plt

# --- Step 1: Generate predictions for both models
y_pred_logreg = log_reg.predict(X_test)
y_pred_xgb = xgb_model.predict(X_test)

# --- Step 2: Define core performance metrics
metrics = {
    "Accuracy": accuracy_score,
    "Precision": precision_score,
    "Recall": recall_score,
    "F1": f1_score
}

# --- Step 3: Evaluate fairness across protected features
results = {}

for model_name, preds in {
    "Logistic Regression": y_pred_logreg,
    "XGBoost": y_pred_xgb
}.items():
    
    print(f"\n🔹 Evaluating {model_name} Fairness Metrics...\n")
    
    model_metrics = {}
    
    for feature_name, feature_values in protected_features.items():
        frame = MetricFrame(
            metrics=metrics,
            y_true=y_test,
            y_pred=preds,
            sensitive_features=feature_values
        )
        
        dp_diff = demographic_parity_difference(y_test, preds, sensitive_features=feature_values)
        eo_diff = equalized_odds_difference(y_test, preds, sensitive_features=feature_values)
        
        print(f"=== {feature_name} ===")
        print("Group-wise metrics:")
        print(frame.by_group.round(3))
        print(f"Demographic Parity Difference: {dp_diff:.3f}")
        print(f"Equalized Odds Difference:     {eo_diff:.3f}\n")
        
        model_metrics[feature_name] = {
            "by_group": frame.by_group,
            "DPD": dp_diff,
            "EOD": eo_diff
        }
    
    results[model_name] = model_metrics

# --- Step 4: Visualize group fairness results
for model_name, model_results in results.items():
    for feature_name, feature_results in model_results.items():
        frame = feature_results["by_group"]
        frame.plot(kind="bar", figsize=(8,5), title=f"{model_name} - Group Fairness for {feature_name}")
        plt.xlabel(feature_name)
        plt.ylabel("Metric Value")
        plt.grid(axis="y", linestyle="--", alpha=0.7)
        plt.legend(loc="lower right")
        plt.tight_layout()
        plt.show()
```
